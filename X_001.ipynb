# **Several ways to create array using numpy**

```python
import numpy as np

# create regular python array
a = np.array([2,3,4])                 # int array
print ("a = ", a)
print("a.dtype = ", a.dtype)
b = np.array([1.2,3.5,5.1])           # float array
print("b = ", b)
print("b.dtype = ", b.dtype)

# function zeros
c= np.zeros((3,5))
print ("with zeros function c = \n", c)

# function ones
d = np.ones((2,3,4), dtype=np.int16)
print("with ones function d=\n", d)

# function empty
e = np.empty((2,3))
print("with empty function e =\n", e)

e = np.empty((2,3), dtype=np.int16)
print("with empty function e =\n", e)
```
